```python
# Run LDR = Standard PCA
mat = scipy.io.loadmat('../../data/graph/MNIST_data.mat')
X = mat['X']
n = X.shape[0]
d = X.shape[1]
Cgt = mat['C'].squeeze()
print(n,d)

nb_pca = 3
[PC,PD,EnPD] = compute_pca(X,nb_pca)
Xldr = PC[:,0]
Yldr = PC[:,1]
Zldr = PC[:,2]
```

```python
# 3D Plot
fig = pylab.figure(2)
ax = Axes3D(fig)
size_vertex_plot = 1
ax.scatter(Xldr, Yldr, Zldr, s=size_vertex_plot*np.ones(n), c=Cgt)
ax.view_init(24,-105)
ax.axis('off')
plt.title('3D Visualization with PCA') 
pyplot.show()
```
